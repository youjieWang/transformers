## 数据映射

```python
def add_prefix(example):
    example["title"] = 'Prefix: ' + example["title"]
    return example
```

```python
prefix_dataset = datasets.map(add_prefix)
prefix_dataset
```

```python
prefix_dataset["train"][: 10]["title"]
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-chinese")
def preprocess_function(example, tokenizer=tokenizer):
    model_inputs = tokenizer(example["content"], max_length=512, truncation=True)
    labels = tokenizer(example["title"], max_length=32, truncation=True)
    # label就是title编码的结果
    model_inputs["labels"] = labels["input_ids"]  # labels["input_ids"] 因为这个tokenizer返回的是一个字典，字典的键就是“input_ids”
    return model_inputs
```

```python
processed_datasets = datasets.map(preprocess_function)
processed_datasets
```

```python
# 数据量比较大的时候可以使用批处理
processed_datasets = datasets.map(preprocess_function, batched=True)
processed_datasets
```

```python
# 如果模型不支持Rust，你还想要快速处理的话，可以使用num_proc进行多进程处理
# 但是有一个问题就是，多线程的处理方式，主线程的数据跟子线程的数据没有办法实现数据共享，因此在实现映射函数的时候需要传递参数，实现数据共享
processed_datasets = datasets.map(preprocess_function, num_proc=4)
processed_datasets
```

```python
# 如果有一些原始字段你不需要的话你可以使用功能remove_columns来删除
processed_datasets = datasets.map(preprocess_function, batched=True, remove_columns=datasets["train"].column_names)
processed_datasets
```

## 通过自定义加载脚本加载数据集

## Dataset with DataCollator

```python
from transformers import DataCollatorWithPadding
```

```python
dataset = load_dataset("csv", data_files="./all_data/ChnSentiCorp_htl_all.csv", split="train")
dataset = dataset.filter(lambda x: x['review'] is not None)
dataset
```

```python
def process_function(examples):
    tokenizer_examples = tokenizer(examples['review'], max_length=128, truncation=True)
    tokenizer_examples['labels'] = examples['label']
    return tokenizer_examples

```

```python
tokenized_dataset = dataset.map(process_function, batched=True, remove_columns=dataset.column_names)
tokenized_dataset
```

```python
print(tokenized_dataset[:3])
```

```python
collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
from torch.utils.data import DataLoader
dl = DataLoader(tokenized_dataset, batch_size=4, shuffle=True, collate_fn=collator)
```

```python
num = 0
for batch in dl:
    print(batch['input_ids'].size())
    num += 1
    if num > 10:
        break
```
